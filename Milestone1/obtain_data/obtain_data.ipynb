### Bringing in the 5 Worldbank dfs through:
#### Pandas read csv into the system
#### Melt the df to make it long
#### Pivot the df to keep the variable (original column data) and, instead, make the variables our columns
#### Write the files to csv to convert from a pivot_table to a generic table (mostly just propagating the data)
#### Loading the new CSVs back in under a new variable to get the new format(only reading the actual data so skipping headers)
#### Loading the first rows into an alternate df to wrangle the headers (these caused memory errors via dtype warnings if included)
#### Combining the second row of the header data (first three columns) into the first
#### Using a for loop to remove periods and underscores from column titles
#### Renaming the columns via the second df
#### Writing this back to CSV over the first CSV created

```python
pop=pd.read_csv("../initial_data/Population-EstimatesData.csv", sep=",")
pop=pop.drop('Indicator Name', axis=1)
test=pd.melt(pop, id_vars=['Country Name', 'Country Code', 'Indicator Code'])
stacked=pd.pivot_table(test, index=['Country Name', 'Country Code', 'variable'], columns='Indicator Code')
stacked.to_csv("../initial_data/Pop_norm.csv")
pop=pd.read_csv("../initial_data/Pop_norm.csv", skiprows=3, header=None)
col_data=pd.read_csv("../initial_data/Pop_norm.csv", skiprows=1, nrows=2, header=None)
col_data.iloc[0,0:3]=col_data.iloc[1,0:3]
for x in range(len(col_data.iloc[0,:])):
     col_data.iloc[0,x]=col_data.iloc[0,x].replace(".","").replace("_","")
pop.columns=col_data.iloc[0,:]
```

```python
spi=pd.read_csv("../initial_data/SPIData.csv", sep=",")
spi=spi.drop('Indicator Name', axis=1)
test=pd.melt(spi, id_vars=['Country Name', 'Country Code', 'Indicator Code'])
stacked=pd.pivot_table(test, index=['Country Name', 'Country Code', 'variable'], columns='Indicator Code')
stacked.to_csv("../initial_data/SPI_norm.csv")
spi=pd.read_csv("../initial_data/SPI_norm.csv", skiprows=3, header=None)
col_data=pd.read_csv("../initial_data/SPI_norm.csv", skiprows=1, nrows=2, header=None)
col_data.iloc[0,0:3]=col_data.iloc[1,0:3]
for x in range(len(col_data.iloc[0,:])):
     col_data.iloc[0,x]=col_data.iloc[0,x].replace(".","").replace("_","")
spi.columns=col_data.iloc[0,:]
```

```python
gen=pd.read_csv("../initial_data/Gender_StatsData.csv", sep=",")
gen=gen.drop('Indicator Name', axis=1)
test=pd.melt(gen, id_vars=['Country Name', 'Country Code', 'Indicator Code'])
stacked=pd.pivot_table(test, index=['Country Name', 'Country Code', 'variable'], columns='Indicator Code')
stacked.to_csv("../initial_data/Gender_norm.csv")
gen=pd.read_csv("../initial_data/Gender_norm.csv", skiprows=3, header=None)
col_data=pd.read_csv("../initial_data/Gender_norm.csv", skiprows=1, nrows=2, header=None)
col_data.iloc[0,0:3]=col_data.iloc[1,0:3]
for x in range(len(col_data.iloc[0,:])):
     col_data.iloc[0,x]=col_data.iloc[0,x].replace(".","").replace("_","")
gen.columns=col_data.iloc[0,:]
```

```python
hnp=pd.read_csv("../initial_data/HNP_StatsData.csv", sep=",")
hnp=hnp.drop('Indicator Name', axis=1)
test=pd.melt(hnp, id_vars=['Country Name', 'Country Code', 'Indicator Code'])
stacked=pd.pivot_table(test, index=['Country Name', 'Country Code', 'variable'], columns='Indicator Code')
stacked.to_csv("../initial_data/HNP_norm.csv")
hnp=pd.read_csv("../initial_data/HNP_norm.csv", skiprows=3, header=None)
col_data=pd.read_csv("../initial_data/HNP_norm.csv", skiprows=1, nrows=2, header=None)
col_data.iloc[0,0:3]=col_data.iloc[1,0:3]
for x in range(len(col_data.iloc[0,:])):
     col_data.iloc[0,x]=col_data.iloc[0,x].replace(".","").replace("_","")
hnp.columns=col_data.iloc[0,:]
```

```python
wdi=pd.read_csv("../initial_data/WDIData.csv", sep=",")
wdi=wdi.drop('Indicator Name', axis=1)
test=pd.melt(wdi, id_vars=['Country Name', 'Country Code', 'Indicator Code'])
stacked=pd.pivot_table(test, index=['Country Name', 'Country Code', 'variable'], columns='Indicator Code')
stacked.to_csv("../initial_data/WDI_norm.csv")
wdi=pd.read_csv("../initial_data/WDI_norm.csv", skiprows=3, header=None)
col_data=pd.read_csv("../initial_data/WDI_norm.csv", skiprows=1, nrows=2, header=None)
col_data.iloc[0,0:3]=col_data.iloc[1,0:3]
for x in range(len(col_data.iloc[0,:])):
     col_data.iloc[0,x]=col_data.iloc[0,x].replace(".","").replace("_","")
wdi.columns=col_data.iloc[0,:]
```

#### Removing common records

```python
spi=spi.iloc[:,~spi.columns.isin(pop.iloc[:,3:].columns)]

hnp=hnp.iloc[:,~hnp.columns.isin(pop.iloc[:,3:].columns)]
hnp=hnp.iloc[:,~hnp.columns.isin(spi.iloc[:,3:].columns)]

gen=gen.iloc[:,~gen.columns.isin(pop.iloc[:,3:].columns)]
gen=gen.iloc[:,~gen.columns.isin(spi.iloc[:,3:].columns)]
gen=gen.iloc[:,~gen.columns.isin(hnp.iloc[:,3:].columns)]

wdi=wdi.iloc[:,~wdi.columns.isin(pop.iloc[:,3:].columns)]
wdi=wdi.iloc[:,~wdi.columns.isin(spi.iloc[:,3:].columns)]
wdi=wdi.iloc[:,~wdi.columns.isin(hnp.iloc[:,3:].columns)]
wdi=wdi.iloc[:,~wdi.columns.isin(gen.iloc[:,3:].columns)]
```

```python
wdi.to_csv("../initial_data/WDI_norm.csv", index=False)
hnp.to_csv("../initial_data/HNP_norm.csv", index=False)
gen.to_csv("../initial_data/Gender_norm.csv", index=False)
spi.to_csv("../initial_data/SPI_norm.csv", index=False)
pop.to_csv("../initial_data/Pop_norm.csv", index=False)
```

### Generating a df for country codes that we're including in data going forward

```python
#Bring in the entire country codes file
countrycodesDF = pd.read_csv("../initial_data/iso-country-codes_regexed.csv")
# Drop first column, which is just index
countrycodesDF = countrycodesDF.drop("Unnamed: 0", axis=1)
#Rename the columns
countrycodesDF.columns = ["short_name", "alpha2_code", "alpha3_code", "numeric_code", "iso3166-2"]
#Using one of our data sources to effect an inner join on the two 
countrycodesFinal=countrycodesDF[countrycodesDF['alpha3_code'].isin(pop['Country Code'])]
#Exporting the pared-down version
countrycodesFinal.to_csv("../initial_data/CountryCodesFinal.csv", index=False)
```

```python
import getpass
mypasswd = getpass.getpass()
#username = input('Enter username: ')
host = 'pgsql.dsa.lan'
database = 'caponl_22g3'
# Then connects to the DB
from sqlalchemy.engine.url import URL
from sqlalchemy import create_engine

# SQLAlchemy Connection Parameters
postgres_db = {'drivername': 'postgres',
               'username': username,
               'password': mypasswd,
               'host': host,
               'database' :database}
engine = create_engine(URL(**postgres_db), echo=True)
connection_string = f"postgres://{username}:{mypasswd}@{host}/{database}"
%load_ext sql
%sql $connection_string 
del mypasswd
```

```python
pop.to_sql('WB_Pop', engine, if_exists='replace')
spi.to_sql('WB_SPI', engine, if_exists='replace')
gen.to_sql('WB_Gender', engine, if_exists='replace')
hnp.to_sql('WB_HNP', engine, if_exists='replace')
wdi.to_sql('WB_WDI', engine, if_exists='replace')
```
