# Correlation Workbook

This notebook is used as an intermediate step to a cool correlation network visualization. The output of this notebook is the input of the notebook that makes that visualization.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../carpentry/completeDF.csv')
```

```python
df = df.drop(['CountryName', 'Year', 'Continent', 'Region1', 'ISO3'], axis=1)

df.head()
```

```python
df_corr = df.corr()
```

```python
df_corr.head()
```

```python
df_corr_abs = df_corr.abs()

df_corr_abs.head()
```

```python
df_nodes = df_corr_abs.where(df_corr != 1, 0)
df_nodes = df_nodes.where(df_nodes <= 0.9, 1)
#df_nodes = df_nodes.where(df_nodes >= -.9, 1)
#df_nodes = df_nodes.where(df_nodes >= 0.9, 1)
df_nodes = df_nodes.where(df_nodes >= 1, 0)
```

```python
df_nodes.head()
```

```python
df_nodes_int = df_nodes.astype(int)

df_nodes_int.head()
```

```python
df_nodes_int.to_csv('df_nodes.csv')
```
