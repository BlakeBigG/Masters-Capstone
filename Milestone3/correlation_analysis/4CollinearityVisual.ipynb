```r
library("dplyr")
library("tidyr")
library("igraph")

df <- read.csv('df_nodes.csv')
```

```r
df_nodes <- subset(df, select = -X)
df_nodes <- as.matrix(df_nodes)
head(df_nodes)
```

```r

gSA <- graph_from_adjacency_matrix(df_nodes, weighted = NULL,  mode = "undirected")
gSA <- simplify(gSA)
gSA <- set_vertex_attr(gSA, "label", value= colnames(df_nodes))
l1 <- layout.kamada.kawai

plot(gSA, layout=l1, vertex.size=1, vertex.label=NA)
```
